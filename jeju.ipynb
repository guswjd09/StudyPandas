### 불필요한 columns제거

```python
df = df.drop(columns=['t_partialpackage','Date_today','o_family','o_Etc'])
df1 = df1.drop(columns=['t_partialpackage','Date_today','o_family','o_Etc'])
df2 = df2.drop(columns=['t_partialpackage','Date_today','o_family','o_Etc'])
df3 = df3.drop(columns=['t_partialpackage','Date_today','o_family','o_Etc'])
```

### 추출한 데이터값을 그래프로 표현하기

```python
path = 'NanumBrush.ttf'
fontprop = fm.FontProperties(fname=path,size=25)
fig = plt.figure(figsize=(20,10))
axes1 = fig.add_subplot(3,2,1)
axes2 = fig.add_subplot(3,2,2)
axes3 = fig.add_subplot(3,2,3)
axes4 = fig.add_subplot(3,2,4)
axes5 = fig.add_subplot(3,2,5)
axes6 = fig.add_subplot(3,2,6)
axes1.bar(sdf1['Date'], sdf1['t_individualtour'])
axes2.bar(sdf1['Date'], sdf1['t_package'])
axes3.bar(sdf1['Date'], sdf1['o_Leisuresports'])
axes4.bar(sdf1['Date'], sdf1['o_business'])
axes5.bar(sdf1['Date'], sdf1['o_vacation'])
axes6.bar(sdf1['Date'], sdf1['o_educationaltour'])
axes1.set_title("개별여행",fontproperties=fontprop)
axes2.set_title("패키지",fontproperties=fontprop)
axes3.set_title("레저스포츠",fontproperties=fontprop)
axes4.set_title("회의 및 업무",fontproperties=fontprop)
axes5.set_title("휴양 및 관람",fontproperties=fontprop)
axes6.set_title("교육여행",fontproperties=fontprop)
fig.suptitle("목적별 제주여행",fontproperties=fontprop)
fig.tight_layout()
```
