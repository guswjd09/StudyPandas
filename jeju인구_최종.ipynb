```python
def get_yx(addr):
    REST_API_KEY = '7e07c632e0ebe5f080eb24728835be78'
    url = 'https://dapi.kakao.com/v2/local/search/keyword.json?category_group_code=PO3&query={}'.format(addr)
    headers = {'Authorization' : 'KakaoAK {}'.format(REST_API_KEY)}
    r = requests.get(url,headers=headers)
    data = r.json()
    try:
        return [data['documents'][0]['y'],data['documents'][0]['x']]
    except:
        return '결과없음'
```

```python
df["위치"] = df["주소"].apply(get_yx)
```

```python
import folium
from folium.plugins import MarkerCluster
```

```python
def display(loc,map):
    if loc[0] != '결과없음':
        folium.Marker(loc[0],tooltip=loc[1]).add_to(map)

map = folium.Map((33.5150737361162,126.526342790758), zoom_start=9)
cluster = MarkerCluster().add_to(map)

df[['위치','읍면동']].apply(lambda x:display(x,map=cluster),axis=1)
map
```
