## merge
* 데이터베이스 join이랑 비슷

```python
person = pd.read_csv('data/survey_person.csv')
site = pd.read_csv('data/survey_site.csv')
survey = pd.read_csv('data/survey_survey.csv')
visited = pd.read_csv('data/survey_visited.csv')
```

```python
# 일부추출 (0,2,6)
visited_subset = visited.loc[[0,2,6]]
visited_subset
```

```python
# pd.merge()
# 데이터프레임.merge()
# site, visited_subset 연결
# 키 값이 같으면 on사용 다르면 left_on, right_on
site.merge(visited_subset, left_on='name', right_on='site') 
```

```python
# person, survey 연결
person.merge(survey, left_on='ident', right_on='person')
```

```python
# visited, site 연결
visited.merge(site, left_on='site', right_on='name')
```

```python
# visited, survey 연결
visited.merge(survey, left_on='ident', right_on='taken')
```

##### left_on = [' ',' ']키가 여러개이면 대괄호를 사용해서 적어준다

```python
ps = person.merge(survey, left_on='ident', right_on='person')
vs = visited.merge(survey, left_on='ident', right_on='taken')
```

```python
ps.head(1)
```

```python
vs.head(1)
```

```python
ps.merge(vs, left_on=['ident','taken','quant','reading'],
         right_on=['person','ident','quant','reading'])
```

##### _x -> 왼쪽에서 데이터가 옴   
##### _y -> 오른쪽에서 데이터가 옴
