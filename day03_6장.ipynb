## ffill
* 누락값이 나타나기 전의 값으로 누락값이 변경
* 6행 누락값 -> 5행의 값을 사용하여 누락값 처리
* 0과 1행은 처음부터 누락값이기 때문에 누락값 그대로 남아있음

```python
ebola.fillna(method='ffill')
```

## bfill
* 누락값이 나타난 이후의 첫번쨰 값으로 양쪽의 누락값이 모두 변경

```python
ebola.fillna(method='bfill') 
```

## interpolate
* 누락값 양쪽에 있는 값을 이용해서 중간값을 구한 다음 누락값 처리

```python
ebola.interpolate()
```
