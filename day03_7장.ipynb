```python
df = billboard.melt(id_vars=['year','artist','track','time','date.entered'], 
               var_name='week',value_name='rating')
```

```python
df.shape
```

```python
df[df.track=='Loser']
```

```python
df.columns
```

```python
songs = df[['year', 'artist', 'track', 'time']]
```

```python
songs.shape
```

```python
# 중복되는 데이터 제거
songs = songs.drop_duplicates()
```

```python
songs
```

```python
songs.reset_index(inplace=True)
```

```python
songs.columns
```

```python
songs.columns = ['id', 'year', 'artist', 'track', 'time']
```

```python
songs.head(2)
```

```python
result = df.merge(songs, on=['year', 'artist', 'track', 'time'])
```

```python
with open('data/raw_data_urls.txt','r')as data_urls:
    for line,url in enumerate(data_urls):
        if line == 5:
            break
        fn = url.split('/')[-1].strip()
        fp = os.path.join('','raw/',fn) # raw폴더에 저장  
        print(fp)
```

```python
with open('data/raw_data_urls.txt','r')as data_urls:
    for line,url in enumerate(data_urls):
        if line == 5:
            break
        fn = url.split('/')[-1].strip()
        fp = os.path.join('','raw/',fn) # raw폴더에 저장  
        urllib.request.urlretrieve(url,fp) # 다운로드   (주소,경로)
```
