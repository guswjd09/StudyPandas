```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
```

```python
data = pd.read_csv('전국야영장2.csv')
data.head(3)
```

```python
data.columns
```

```python
data.drop(columns = ['일반야영장수', '자동차야영장', '글램핑','카라반', '화장실', '샤워실', '개수대', '소화기', '방화수', '방화사', '화재감지기', 
                     '기타부대시설 1','기타부대시설 2', '데이터기준일자', '인허가일자'])
```

```python
map = folium.Map((35.1607753,129.0472984),zoom_start = 10)

coords = data[['위도', '경도']]

marker_cluster = MarkerCluster().add_to(map)

for item in data.iterrows():
    location = ([item[1]['위도'],item[1]['경도']])
    name = item[1]['캠핑(야영)장명']
    folium.Marker(location,tooltip=name).add_to(marker_cluster)

map
```
